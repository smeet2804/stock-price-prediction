```python
import numpy as np
import pandas as pd
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
import matplotlib.pyplot as mlpt
import csv
import random
```

```python
cdata = pd.read_csv("/content/INFYFINALSENT.csv")
ccdata = pd.read_csv("/content/INFYFINALSENT.csv")
ccdata
```

```python
indx=0
get_tweet=""
for i in range(0,len(cdata)-1):
    get_date=cdata.Date.iloc[i]
    next_date=cdata.Date.iloc[i+1]
    if(str(get_date)==str(next_date)):
        get_tweet=get_tweet+cdata.Summary.iloc[i]+" "
    if(str(get_date)!=str(next_date)):
        ccdata.at[indx,'Date']=get_date
        ccdata.at[indx,'Summary']=get_tweet
        indx=indx+1
        get_tweet=" "
```

```python
ccdata = ccdata.groupby(['Date'])['Summary'].apply(list)
```

```python
ccdata.to_csv("Merged.csv")
```

```python
df_merged=pd.read_csv('/content/Tech_Sent.csv')
df_merged
```

```python
df_merged=df_merged[['Date','Open','High','Low','Close','Volume','Summary']]
df_merged.head()
```

```python
df_merged['compund']=''
df_merged['neg']=''
df_merged['neu']=''
df_merged['pos']=''

df_merged.head()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia=SentimentIntensityAnalyzer()
```

```python
for index,sentence in enumerate(df_merged['Summary']):
    ps=sia.polarity_scores(sentence)
    df_merged['compund'][index]=ps['compound']
    df_merged['neg'][index]=ps['neg']
    df_merged['neu'][index]=ps['neu']
    df_merged['pos'][index]=ps['pos']
```

```python
final_df = df_merged[['Date','compund','neg','neu','pos','Open','High','Low','Close','Volume']]
```

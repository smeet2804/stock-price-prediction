```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df=pd.read_csv('/content/INFY_Final_CNN.csv')
df
```

```python
pos = df['pos']
neg=df['neg']
neu=df['neu']
compound=df['compound']
close = df['close price']

corr_pos, _ = pearsonr(close, pos)
corr_neg, _ = pearsonr(close, neg)
corr_neu, _ = pearsonr(close, neu)
corr_compound, _ = pearsonr(close, compound)

print('Pearsons correlation: %.3f' % corr_pos)
print('Pearsons correlation: %.3f' % corr_neg)
print('Pearsons correlation: %.3f' % corr_neu)
print('Pearsons correlation: %.3f' % corr_compound)
```
